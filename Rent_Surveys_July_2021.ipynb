```python
# import urllib library
from urllib.request import urlopen
import json
import pandas as pd
import geopandas
```

```python
Prices=list(range(0, 3000,25))
```

```python

#Building_ Urls
List_urls=[]
for i in range(0,len(Prices)):
    new_url="https://www.rentfaster.ca/api/map.json?e=undefined&beds=&baths=&type=&price_range_adv%5Bfrom%5D="+ str(Prices[i])+ "&price_range_adv%5Bto%5D="+str(Prices[i]+24)+"&area=51.18725531061486%2C-113.72722341074217%2C50.90397819717543%2C-114.3877763892578"
    List_urls.append(new_url)
  
```

```python
#Transforming all the listings into a single dataframe we have to crawl over all URLS and 
def get_listings(List_urls):
    list_of_properties=[]
    for i in range(0,len(List_urls)):
        response = urlopen(List_urls[i])
        data_json = json.loads(response.read())
        if len(data_json['listings'])!=0:
            new_dataframe = pd.json_normalize(data_json['listings'])
            list_of_properties.append(new_dataframe)
            print("Added to the list of databases")
        else:    
            print("empty price range")
    large_df = pd.concat(list_of_properties, ignore_index=True)
    return large_df
```

```python
# Getting all the properties
final=get_listings(List_urls)
```

```python
final.to_csv('Rent_Faster_July_2021_1.csv')
```

# DETAILED LISTING DATA

```python
#Building_Property urls
property_ids=list(final["id"])
#Loop
List_properties_urls=[]
for i in range(0,len(property_ids)):
    new_url="https://www.rentfaster.ca/api/"+str(property_ids[i])+".json"
    List_properties_urls.append(new_url) 
```

```python
final_Detail=get_listings(List_properties_urls)
```

# Making a Color Coded Map with Calgary Communities

```python
#Load data both Rent faster Data anad Shapefiles file
Rent_Faster_data=pd.read_csv('Rent_Faster_July_2021_1.csv')
Calgary_comunities = geopandas.read_file("./Community Boundaries/geo_export_ed6f5719-88f3-4c34-bd13-9e7416a438dc.shp") 

#Extra Formating
Rent_Faster_data['community'] = Rent_Faster_data['community'].str.upper()
Calgary_comunities=Calgary_comunities.rename(columns={'name': 'community'})
```

```python
#Crosstab, average rent Prices by community
Price_Community=pd.DataFrame(Rent_Faster_data.groupby('community')['price'].mean())
Price_Community
```

```python
#Average Rent Prices by Community and type 
Price_Community_House_type=pd.DataFrame(Rent_Faster_data.groupby(['community', 'type'])['price'].mean())
Price_Community_House_type
```

```python
#Join Community Average Overall
Calgary_comunities_overall=Calgary_comunities.merge(Price_Community, how='inner', on='community')
```

```python
#Columns
Rent_Faster_data['type'].unique()
```

```python
#Single Detach Homes Rent Prices
# Plot population estimates with an accurate legend
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200

Rent_Faster_data_House=Rent_Faster_data[Rent_Faster_data['type']=="['Apartment']"]
Merged_Data_Houses=Calgary_comunities.merge(Rent_Faster_data_House, how='left', on='community')

#Map
fig, ax = plt.subplots(1, 1)
Merged_Data_Houses.plot(column='price', ax=ax, legend=True,missing_kwds= dict(color = "lightgrey",))
```
